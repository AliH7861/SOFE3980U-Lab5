##3. Load Labels.csv

Download and upload the [Labels.csv](https://github.com/zubxxr/SOFE3980U-Lab5/blob/main/Labels.csv) into this notebook, and then load the file.

```python
labels_url = "https://raw.githubusercontent.com/AliH7861/SOFE3980U-Lab5/main/Labels.csv"

try:
    df = pd.read_csv("Labels.csv")
    print("Loaded Labels.csv from the notebook folder.")
except FileNotFoundError:
    df = pd.read_csv(labels_url)
    print("Loaded Labels.csv from GitHub.")

print(df.shape)
df.head()
```

##4. Preview the Dataset

```python
df.head()
```

##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context(mode="ephemeral")
data_source = context.data_sources.add_pandas("pandas_labels")
data_asset = data_source.add_dataframe_asset(name="labels_asset")
```

##6. Define and Create a Data Batch

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe("labels_batch")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

##7. Define Three Expectations for Column Values

Using this [link](https://greatexpectations.io/expectations/), choose three expectation functions and apply them to the labels dataset in a relevant manner.

You should replace the 'ExpectColumnValuesToBeBetween' function with other functions you select from the link.

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Timestamp"
)
expectation_1
```

### Validate Data Against Expectation 1

```python
validation_result_1 = batch.validate(expectation_1)
print(validation_result_1)
```

### Expectation 2

```python
expectation_2 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Occluded_Image_view",
    regex=r"^A_\d{3}\.png$"
)
expectation_2
```

### Validate Data Against Expectation 2

```python
validation_result_2 = batch.validate(expectation_2)
print(validation_result_2)
```

### Expectation 3

```python
expectation_3 = gx.expectations.ExpectColumnPairValuesAToBeGreaterThanB(
    column_A="pedestrianLocationX_BottomRight",
    column_B="pedestrianLocationX_TopLeft"
)
expectation_3
```

### Validate Data Against Expectation 3

```python
validation_result_3 = batch.validate(expectation_3)
print(validation_result_3)
```
